```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import math
%matplotlib inline

n = 10000

theta = math.pi/4.0
P = np.matrix([[math.cos(theta),-math.sin(theta)],[math.sin(theta),math.cos(theta)]])
D = np.matrix([[10.0,0.0],[0.0,1.0]])
Pt = np.transpose(P)
S = P.dot(D.dot(Pt))
print(S)
# example data
X = np.random.randn(n)
Y = np.random.randn(n)
plt.scatter(X,Y)
plt.show()

L = np.linalg.cholesky(S)
print(L)

for i in range(n):
    Z = np.random.randn(2)
    v = L.dot(Z)
    X[i] = v[0,0]
    Y[i] = v[0,1]

plt.scatter(X,Y)
```

```python

num_bins = 50
# the histogram of the data
n, bins, patches = plt.hist(X, num_bins, normed=1, facecolor='green', alpha=0.5)
# add a 'best fit' line
mu = np.mean(X)
sigma = np.std(X)
print(mu,sigma)
b = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, b, 'r--')
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'Histogram of X')
plt.show()

num_bins = 50
# the histogram of the data
n, bins, patches = plt.hist(Y, num_bins, normed=1, facecolor='green', alpha=0.5)
# add a 'best fit' line
mu = np.mean(Y)
sigma = np.std(Y)
print(mu,sigma)
b = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, b, 'r--')
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'Histogram of Y')
plt.show()

C = np.cov(X,Y)
print(C)
```

```python
def Bernoulli(n,p):
    B = np.zeros(n)
    U = np.random.uniform(0.0,1.0,n)
    for i in range(n):
        if (U[i] < 1-p):
            B[i] = 0
        else:
            B[i] = 1
    return B

B = Bernoulli(10000,0.1)
print(B)
print(sum(B))
Bmean = np.mean(B)
Bstd  = np.std(B)
Bvar  = np.var(B)
print(Bmean,Bstd,Bvar)
```
